```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
from datasets import load_dataset
import numpy as np

dataset = load_dataset('lukebarousse/data_jobs')

df = pd.DataFrame(dataset['train'])

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'] )
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DE = df[df['job_title_short']=='Data Engineer'].copy()
```

```python
df_DE
```

```python
df_DE['job_posted_monthno'] = df_DE['job_posted_date'].dt.month
```

```python
df_DE['job_posted_monthno']
```

```python
df_DE_explode = df_DE.explode('job_skills')
```

```python
df_DE_explode
```

```python
df_DE_pivot = df_DE_explode.pivot_table(index='job_posted_monthno', columns='job_skills', aggfunc='size' , fill_value=0)
```

```python
df_DE_pivot.loc['Total']= df_DE_pivot.sum()
```

```python
df_DE_pivot
```

```python
df_DE_pivot = df_DE_pivot[df_DE_pivot.loc['Total'].sort_values(ascending=False).index]

df_DE_pivot.drop('Total', inplace=True)
```

```python
df_DE_pivot
```

```python
df_DE_pivot.iloc[:, :5].plot(kind='line', stacked=True, figsize=(12, 7))
```

```python
df_DE_pivot = df_DE_pivot.reset_index()
```
