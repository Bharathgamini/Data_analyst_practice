```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datasets import load_dataset
import ast


dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()


df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else x)
```

```python
df.head()
```

```python
df['job_work_from_home'].value_counts().plot(kind='bar')
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie')
```

```python
df_pie = df.groupby(['job_work_from_home', 'job_title_short']).size()
```

```python
fig, ax = plt.subplots()
df_pie.plot(kind='pie', subplots=True, startangle=90, autopct='%1.1f%%', ax=ax)
plt.title('Work From Home Distribution by Job Title')
plt.ylabel('')
plt.tight_layout()
```

```python
df['job_title_short'].value_counts().plot(kind='pie')
plt.title('Distribution of Job Titles')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.pie(df['job_title_short'].value_counts(), labels=df['job_title_short'].value_ counts().index, autopct='%1.1f%%', startangle=90)
plt.show()
```
