Aggregate skill counts monthly
re analyst based on percentage of total jobs
plot the monthly skill demand

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import ast
from datasets import load_dataset
from datetime import date, datetime


dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills']= df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_US = df[(df['job_title']=='Data Analyst') &(df['job_country']=='United States')].copy()

df_DA_US['job_posted_monthno']=df_DA_US['job_posted_date'].dt.month

df_DA_US['job_posted_monthno']
```

```python
df_DA_USexplode = df_DA_US.explode('job_skills')
```

```python
df_DA_US_pivot = df_DA_USexplode.pivot_table(index='job_posted_monthno', columns = 'job_skills', aggfunc = 'size', fill_value =0)

df_DA_US_pivot.loc['Total']= df_DA_US_pivot.sum()

df_DA_US_pivot=df_DA_US_pivot[df_DA_US_pivot.loc['Total'].sort_values(ascending=False).index]
```

```python
df_DA_US_pivot.loc['Total'].sort_values(ascending=False)
```

```python
df_DA_US_pivot=df_DA_US_pivot.drop('Total')
```

```python
df_DA_US_pivot
```

```python
df_DA_US_pivot.plot(kind='line')
```

```python
df_DA_Total = df_DA_US.groupby('job_posted_monthno').size()
```

```python
df_DA_Total
```

```python
df_DA_percent = df_DA_US_pivot.div(df_DA_Total/100, axis =0)
df_DA_percent
```

```python
df_DA_percent=df_DA_percent.reset_index()
df_DA_percent['job_posted_month'] = df_DA_percent['job_posted_monthno'].apply(lambda x: pd.to_datetime(x, format = '%m').strftime('%b'))
df_DA_percent=df_DA_percent.set_index('job_posted_month')
```

```python
df_DA_percent = df_DA_percent.drop(columns='job_posted_monthno')
```

```python
df_DA_percent
```

```python
df_plot = df_DA_percent.iloc[:, :5]

sns.lineplot(df_plot, dashes = False, palette = 'tab10')
sns.set_theme(style='ticks')
sns.despine()
plt.title('Percentage of job per month')
plt.ylabel('likelihood in job posting')
plt.xlabel('2023')
plt.legend().remove()
ax = plt.gca()
from matplotlib.ticker import PercentFormatter
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))
for i in range(5):
    plt.text(11.5, df_plot.iloc[-1, i], df_plot.columns[i])
plt.show()
```
